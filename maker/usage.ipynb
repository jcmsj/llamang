```python
# Using ocr.py

import json
from ocr import fits_to_imgs, plot_boxes, to_dict, update_rects_to_actual_textbox

pdf_path = r"D:\datasets\uspto_generated_forms\aia0067_Power to Inspect_Copy - For Applications Filed on _flat_20250312_143650_var46.pdf"
adjusted = update_rects_to_actual_textbox(pdf_path)
images = fits_to_imgs(pdf_path)

# json
print(json.dumps(to_dict(images, adjusted, pdf_path),indent=2))
```
